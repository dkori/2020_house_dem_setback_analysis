```python
# read in beautiful soup and pandas
from bs4 import BeautifulSoup
import requests
import pandas as pd
import regex as re
```

```python
url = "https://en.wikipedia.org/wiki/Congressional_Progressive_Caucus#House_members"
req = requests.get(url)
soup = BeautifulSoup(req.content, 'html.parser')
print(soup.prettify())
```

```python
member_list = soup.find('div','div-col columns column-width').find_all('li')
districts = [x.get_text() for x in member_list]
member_names = [re.sub(' \(.*','',x) for x in districts]
districts2 = [re.sub('.* \(','',x) for x in districts]
districts3 = [re.sub('\)|,.*','',x) for x in districts2]
districts3
df = pd.DataFrame({"member_name" : member_names,
                 "district" : districts3})
df.to_csv('progressive_caucus_members_29_nov_2020.csv',
         index = False)
```
