```python
# read in beautiful soup and pandas
from bs4 import BeautifulSoup
import requests
import pandas as pd
import regex as re
```

```python
url = "https://en.wikipedia.org/w/index.php?title=Cook_Partisan_Voting_Index&oldid=991678053"
req = requests.get(url)
soup = BeautifulSoup(req.content, 'html.parser')
print(soup.prettify())
```

```python
#find the cd table (second table in the page)
cd_table = soup.find_all("table")[1]
```

```python
# get column names from table header
header_soup = cd_table.find_all("th")
column_names = []
for x in header_soup:
    column_names.append(re.sub('\n','',x.get_text()))
column_names
# store list of rows
body_soup = cd_table.find_all('tr')[1:]
# parse list of rows to create list of lists
table_body_list = []
for x in body_soup:
    datapoints = x.find_all('td')
    temp_list = [datapoints[0].get_text(),datapoints[1].get_text(),datapoints[2].get_text()]
    temp_list = [re.sub('\n','',y) for y in temp_list]
    table_body_list.append(temp_list)
# place into pandas dataframe
df = pd.DataFrame(table_body_list,
                 columns = column_names)
```

```python
# save dataframe
df.to_csv("cook index by district.csv")
```
